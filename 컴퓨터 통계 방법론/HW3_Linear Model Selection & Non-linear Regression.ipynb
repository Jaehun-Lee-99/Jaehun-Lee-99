```python
X = np.random.uniform(0,2*math.pi,50)
X_range = np.linspace(0,2*math.pi,1000)
ground_X = np.sin(X_range)
g_X = np.sin(X)
e = np.random.normal(0,0.2,50)
Y = g_X + e
plt.scatter(X,Y,label='data')
plt.plot(X_range, ground_X, 'r', label='ground truth')
plt.show
plt.legend()
```

```python
plt.scatter(X,Y,label='data')
plt.scatter(X_range,0*X_range,s=0.3,c='g',label='zero reg')
plt.show
plt.legend()
```

```python
g1 = np.polyfit(X,Y,0)
plt.scatter(X,Y,label='data')
plt.scatter(X_range,X_range*0+g1,s=0.3,c='g',label='const reg')
plt.show
plt.legend()
print(g1)
```

```python
g2 = np.polyfit(X,Y,1)
plt.scatter(X,Y,label='data')
plt.scatter(X_range,g2[1]+g2[0]*X_range,s=0.3,c='g',label='lin reg')
plt.show
plt.legend()
```

```python
g3 = np.polyfit(X,Y,2)
plt.scatter(X,Y,label='data')
plt.scatter(X_range,g3[2]+g3[1]*X_range+g3[0]*X_range**2,s=0.3,c='g',label='quad reg')
plt.show
plt.legend()
```

```python
X = np.random.uniform(0,2*math.pi,10)
g_X = np.sin(10*X)
X_range = np.linspace(0,2*math.pi,10000)
e = np.random.normal(0,0.2,10)
Y = g_X + e
g4 = np.polyfit(X,Y,10)
plt.scatter(X,Y,label='data')
Y_itp = 0*X_range
for i in range(11):
  Y_itp += g4[10-i]*X_range**(i)
plt.scatter(X_range,Y_itp,s=0.1,c='g',label='interpolation')
plt.ylim([-5,5])
plt.show
plt.legend()
```

```python
df = pd.read_csv('/content/drive/MyDrive/ISLP/College.csv')
df1 = pd.get_dummies(df.iloc[:,1:])
df = pd.concat([df.iloc[:,0],df1],axis=1)
df.head()
```

```python
X = df.iloc[:, df.columns != 'Outstate']
y = df['Outstate']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=1)
```

```python
X.head()
```

```python
lr = LinearRegression()

from mlxtend.feature_selection import SequentialFeatureSelector as SFS
from mlxtend.plotting import plot_sequential_feature_selection as plot_sfs
sfs = SFS(lr,
          k_features = 18,  # We have 18 features
          forward = True,
          floating = False,
          scoring = 'r2',
          cv = 0)

sfs = sfs.fit(X_train.iloc[:,1:], y_train)  # as_matrix() to be readable by sfs

fig = plot_sfs(sfs.get_metric_dict())

#plt.title('Sequential forward selection (w. StdDev)')
plt.title('Sequential forward selection')
plt.grid()
plt.show()
```
